```python
#Q1
def smart_parking_system(capacity, logs):
    current_parked = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1
        
        if current_parked > peak_usage:
            peak_usage = current_parked
    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)
    if current_parked > capacity:
        print("ALERT: Parking capacity exceeded!")
    else:
        print("Parking Status: Available")
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

```python
#Q2
def estimate_delivery_time(distance, traffic, weather):
    speed = 30  # km per hour
    prep_time = 20  # minutes
    
    travel_time = (distance / speed) * 60
    
    if traffic == "Low":
        traffic_delay = 5
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 20
    else:
        traffic_delay = 0

    if weather == "Clear":
        weather_delay = 0
    elif weather == "Cloudy":
        weather_delay = 5
    elif weather == "Rainy":
        weather_delay = 15
    elif weather == "Storm":
        weather_delay = 25
    else:
        weather_delay = 0
    
    total_time = prep_time + travel_time + traffic_delay + weather_delay
    
    print("Estimated Delivery Time:", round(total_time), "minutes")
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

```python
#Q3
def seat_occupancy_analyzer(total_seats, booked_seats):
    
    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", round(occupancy), "%")
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
    if occupancy >= 90:
        print("Suggestion: Consider opening an additional show.")
total_seats = 200
booked_seats = [1] * 150   # Simulating 150 booked seats

seat_occupancy_analyzer(total_seats, booked_seats)
```

```python
#Q4
def classify_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)
    
    print("Average CPU Load:", round(average), "%")
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
cpu_data = [45, 60, 70, 85, 90]

classify_server_load(cpu_data)
```

```python
#Q5
def classroom_resource_monitor(resource_usage):
    
    overused = []
    threshold = 8  
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
resource_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resource_data)
```

```python
#Q6
def event_registration_controller(capacity, total_registrations):
    
    confirmed = 0
    waitlisted = 0
    
    for i in range(1, total_registrations + 1):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

event_capacity = 100
registrations = 105

event_registration_controller(event_capacity, registrations)
```
