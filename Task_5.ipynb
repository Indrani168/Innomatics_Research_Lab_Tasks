```python
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example usage
likes = [200, 150, 300, 400]   # Total = 1050
analyze_engagement(likes)
```

```python
def analyze_rainfall(rainfall_data, required_level):
    
    total = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total += rain
    
    # Calculate average rainfall
    average = total / len(rainfall_data)
    
    # Check rainfall status
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)


# Example usage
rainfall = [70, 75, 80, 63]   # Average = 72
required = 70
analyze_rainfall(rainfall, required)
```

```python
def check_duplicate_usernames(usernames):
    
    unique_users = set()
    duplicate_found = False
    
    for user in usernames:
        if user in unique_users:
            duplicate_found = True
            break
        unique_users.add(user)
    
    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
        print("Status: Flag accounts")
    else:
        print("Duplicate Accounts Found: No")
        print("Status: Accounts are unique")


# Example usage
users = ["rahul123", "neha45", "amit99", "rahul123"]
check_duplicate_usernames(users)
```

```python
def filter_premium_crops(prices):
    
    premium_crops = []
    
    # Loop to filter premium crops
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)


# Example usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
def evaluate_application_health(error_count):
    
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", error_count)
    print("System Status:", status)


# Example usage
evaluate_application_health(7)
```

```python
def validate_transaction(amount):
    
    daily_limit = 50000
    
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example usage
validate_transaction(60000)
```

```python
def check_exam_eligibility(attendance_list):
    
    total_classes = len(attendance_list)
    present_count = 0
    
    # Loop to count present days
    for status in attendance_list:
        if status == "P":
            present_count += 1
    
    # Calculate attendance percentage
    percentage = (present_count / total_classes) * 100
    
    # Check eligibility
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


# Example usage
attendance = ["P", "P", "A", "P", "P"]  # 4 out of 5 present → 80%
check_exam_eligibility(attendance)
```

```python
def calculate_electricity_bill(units):
    
    bill = 0
    
    # Billing calculation using slab logic
    if units <= 100:
        bill = units * 3
    
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    return bill, usage_status


# Example usage
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Classification:", status)
```
