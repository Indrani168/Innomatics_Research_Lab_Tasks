```python
#Q1

sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to get unique words
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#Q2
# Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

# Printing result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
#Q3
# List of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
#Q4
# List of product prices
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
#Q5
# Attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count Present days
present_count = attendance.count("P")

# Total days
total_days = len(attendance)

# Calculate percentage
percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python
#Q6

phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
#Q7

text = "python"

# Create empty dictionary
char_count = {}

# Loop through each character
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```

```python
#Q7

numbers = [10, 20, 30]

# Convert list to tuple
numbers_tuple = tuple(numbers)

print("Tuple:", numbers_tuple)
```

```python
#Q9
# Sample dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key_name = "Ravi"

# Check if key exists
if key_name in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
#Q10

marks = [70, 75, 80, 65]

# Calculate sum of marks
total = sum(marks)

# Calculate average
average = total / len(marks)

print("Average Marks:", average)
```
